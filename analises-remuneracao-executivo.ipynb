O presente trabalho avalia a evolução da remuneração média mensal no Executivo civil federal ativa, em função de sexo e raça, entre os anos de 1999 e 2020, a partir de dados disponibilizados pelo Ipea nesse sentido.

```python
import requests
import pandas as pd

#coleta dos dados
with open("dados.csv", "w") as f:
    f.write(requests.get("https://www.ipea.gov.br/atlasestado/arquivos/downloads/5233-liquidosexoraca.csv").text)
    
dados = pd.read_csv("dados.csv", sep=";", decimal=",")
```

```python
#tratamento dos dados
dados.drop("Unnamed: 0", axis=1, inplace=True)
dados_hb = dados.loc[dados["sexo_raça"] == "Homem Branco"]
dados_hn = dados.loc[dados["sexo_raça"] == "Homem Negro"]
dados_mn = dados.loc[dados["sexo_raça"] == "Mulher Negra"]
dados_mb = dados.loc[dados["sexo_raça"] == "Mulher Branca"]
```

```python
import matplotlib.pyplot as plt

#plot
plt.plot(dados_hb["ano"], dados_hb["liquido"], label = "Homem Branco")
plt.plot(dados_hn["ano"], dados_hn["liquido"], label = "Homem Negro")
plt.plot(dados_mn["ano"], dados_mn["liquido"], label = "Mulher Negra")
plt.plot(dados_mb["ano"], dados_mb["liquido"], label = "Mulher Branca")
plt.legend()
plt.show()
```

### Tendências observadas:
Uma das tendências observadas nos dados analisados acima é a de que a média líquida da remuneração percebida pelas mulheres é inferior à média dos homens, fenômeno observado em todos os anos. Ainda, essa ocorrência é mais perceptível nos indivíduos brancos do que nos negros. Esse contraste entre os gêneros pode ser resultado de desigualdade nos cargos ou funções ocupados por homens e mulheres, com eventual presença majoritária de homens em posições de chefia nos órgãos, constatação que demandaria análises complementares.

Além disso, principalmente nos anos de 2009 e 2010, é possível perceber movimento de intensificação das diferenças que já existiam entre as médias das remunerações líquidas dos servidores de cada uma das raças. Ou seja, a partir desses anos, as desigualdades já existentes entre servidores públicos brancos e negros na remuneração percebida se tornou ainda maior. A compreensão desse fenômeno dependeria de análises complementares, mas é possível que seja resultado de reajustes concedidos a carreiras ocupadas majoritariamente por servidores brancos, ou ainda de reajustes diferentes concedidos a cada carreira. Um ato normativo que possivelmente teve impacto nesse fenômeno foi a [Lei n.º 11.907/2009]( 
http://www.planalto.gov.br/ccivil_03/_ato2007-2010/2009/lei/l11907.htm
), derivada da MPV 441/2008 e que promoveu uma série de reestruturação de carreiras do Executivo federal.
